```python
# 
import numpy as np               # 
import matplotlib.pyplot as plt  # 

#  
def sigmoid(x):
    return 1/(1+np.exp(-x))

#  
data_set=np.loadtxt('training.txt')

#   
np.random.shuffle(data_set) 

#   
x=data_set[:,0:2]  #  1000,  2(0~1) #(1000,2)
t=data_set[:,2:3]  #  1000,  1(2) #(1000,1)

#layer  (bias=0)
K=2
N=8
M=1

#learning_rate 
lr=0.01

#    
w=np.random.random((K,N)) #input~hidden layer          # (2, 8)
wp=np.random.random((N,M)) #hidden layer~output layer  # (8, 1)

#  
num_epoch=2000

#forward
EPOCH=[]
LOSS=[]
for epoch in range(num_epoch): 
    E = 0
    EPOCH.append(epoch+1)    #EPOCH  epoch  
    for e in range(0, 1000):
        u=np.dot(x[e], w)    #(8,)
        h=sigmoid(u)         #(8,)
        up=np.dot(h, wp)    # (1,)
        y=sigmoid(up)       # (1,)

    #loss function
        E_temp=(1/2)*(y-t[e])**2 #(1,)
        E += E_temp[0]           #E 1000   
         
    #back propagation
        EIP = (y-t[e])*y*(1-y) #(1,)     # wp=(8,1)
        EI = np.sum(EIP*wp)*h*(1-h)      #(8,)       
        EI = np.expand_dims(EI, axis=1)  #(8,1)
        
        dwp = EIP * h            # (8, 1)
        dw = (EI*x[e]).T        # (2,8) = ((8,1) * (1,2)).T #  .T
        
        # weight update
        wp = wp - lr*dwp      
        w = w - lr*dw    
    LOSS.append(E/1000)         # LOSS  E/1000  
   
      
    print("epoch: {}, train_loss: {} " .format(epoch+1, E/1000)) # 
                   
    
# 

plt.plot(EPOCH,LOSS)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()   

# 
#1)epoch   LOSS    .
#2)epoch   ( 500) epoch max  LOSS  .
   #   , epoch   LOSS        

#
#1) , FORWARD BACK PROPAGATION  (W,WP)   
#2) ,   /      

```
